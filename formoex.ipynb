```python
def plot_to_file_for_company(c):
    moex = MOEX()
    data = moex.history_engines_stock_totals_securities(date_start='2018-01-01', date_end = '2018-05-01', secid=[c])
    data[["SYSTIME", "SECID", "OPEN", "CLOSE", "LOW", "HIGH", "VOLUME"]]
    data = data[["SYSTIME", "SECID", "OPEN", "CLOSE", "LOW", "HIGH", "VOLUME"]]
    data = data.rename(columns={'SYSTIME':'DATE','VOLUME':'VOL'})
    sber2 = data[data['SECID'] == c]
    sber = data
    plt.plot(sber['VOL'].astype(float), label ='VOL')
    plt.legend()
    plt.savefig('volotdelno'+'.png',format = 'png',dpi=1000)
    plt.show
    
    plt.plot(sber['OPEN'].astype(float),label = 'OPEN')
    plt.plot(sber['LOW'].astype(float),label = 'LOW')
    plt.plot(sber['HIGH'].astype(float),label = 'HIGH')
    plt.plot(sber['CLOSE'].astype(float),label = 'CLOSE')
    plt.savefig('nameofgrapic'+'.png', format='png', dpi=100)
    plt.legend()
    plt.show()
```

```python
plot_to_file_for_company('SBER')
```

```python
def plot_for_companies(d=None, filename=None):
    moex = MOEX()
    data = moex.history_engines_stock_totals_securities(date_start='2018-01-01', date_end = '2018-05-01', secid=[d])
    data[["SYSTIME", "SECID", "OPEN", "CLOSE", "LOW", "HIGH", "VOLUME"]]
    data = data[["SYSTIME", "SECID", "OPEN", "CLOSE", "LOW", "HIGH", "VOLUME"]]
    data = data.rename(columns={'SYSTIME':'DATE','VOLUME':'VOL'})
    data1 = data[data['SECID'] == d]
    
    plt.plot(data['VOL'].astype(float), label = 'VOL')
    plt.legend()
    plt.savefig(filename+'1'+'.png',format='png',dpi=100)
    plt.show()
    plt.plot(data['OPEN'].astype(float), label = 'OPEN')
    plt.plot(data['LOW'].astype(float) ,label ='LOW' )
    plt.plot(data['HIGH'].astype(float) ,label = 'HIGHT')
    plt.plot(data['CLOSE'].astype(float),label = 'CLOSE')
    plt.legend()
    if filename == None:
         plt.show()
    else:
        plt.savefig(filename+'.png', format='png', dpi=100)
        plt.show()
```

```python
def filter_by_date(start,end):
    moex = MOEX()
    data = moex.history_engines_stock_totals_securities(date_start=start, date_end =end , secid=['SBER'])
    data[["SYSTIME", "SECID", "OPEN", "CLOSE", "LOW", "HIGH", "VOLUME"]]
    data = data[["SYSTIME", "SECID", "OPEN", "CLOSE", "LOW", "HIGH", "VOLUME"]]
    data = data.rename(columns={'SYSTIME':'DATE','VOLUME':'VOL'})
    sber2 = data[data['SECID'] == 'SBER' ]
    sber = data
    plt.plot(sber['VOL'].astype(float), label ='VOL')
    plt.legend()
    plt.show()
    
    plt.plot(sber['OPEN'].astype(float), label = 'OPEN')
    plt.plot(sber['LOW'].astype(float), label = 'LOW')
    plt.plot(sber['HIGH'].astype(float), label = 'HIGH')
    plt.plot(sber['CLOSE'].astype(float), label = 'CLOSE')
    plt.legend()
    plt.show
```
